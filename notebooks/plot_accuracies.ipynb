```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#df = pd.read_csv('notebooks/cached_data/accuracy.csv')
df = pd.read_csv("C:/Users/jeand/OneDrive/Bureau/Clean_PC/neuroscience project/accuracy.csv")
```

```python
df.head()
```

```python
#get rid of rows with delay < 0.25

df = df[df['delay'] >= 0.25]
df.head()
```

```python
plt.figure(figsize=(10, 6))

for area in df["area"].unique():
    mask = df["area"] == area
    plt.plot(df[mask]["delay"], df[mask]["accuracy"], label=area)

plt.xlabel("Delay (s)")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy of the SVM classifier for different delays")
plt.legend()
```
